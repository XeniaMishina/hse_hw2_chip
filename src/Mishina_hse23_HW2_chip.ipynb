## Подсчёт ngsplot при помощи deeptools
Скачиваем аннотацию 
Для hg38 - https://hgdownload-test.gi.ucsc.edu/goldenPath/hg38/bigZips/genes/

```python
!wget https://hgdownload-test.gi.ucsc.edu/goldenPath/hg38/bigZips/genes/hg38.knownGene.gtf.gz
```

Файл ENCFF486POV (alignments), соответствующий ENCFF906ZAG (fold change over control)

```python
!wget https://www.encodeproject.org/files/ENCFF906ZAG/@@download/ENCFF906ZAG.bigWig
```

```python
# подсчёт матрицы. В примере считается 30 минут.
%%time
!computeMatrix scale-regions \
 -S ENCFF906ZAG.bigWig \
 -R hg38.knownGene.gtf.gz \
 -out matrix.tab.gz \
 -a 2000 -b 2000 \
 --regionBodyLength 4000 \
 --skipZeros \
 --missingDataAsZero
```

```python
!plotHeatmap \
 -m matrix.tab.gz \
 -out result.png \
 --colorMap YlGnBu \
 --regionsLabel 'genes' \
 --heatmapHeight 15 \
 --plotTitle 'ngs_plot' &
```
